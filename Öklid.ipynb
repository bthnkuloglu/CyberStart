```python
import math

# Öklid mesafesi hesaplayan fonksiyon
def euclideanDistance(point1, point2):
    return math.sqrt((point2[0] - point1[0])**2 + (point2[1] - point1[1])**2)

points = [(1, 2), (4, 6), (7, 8), (2, 1), (9, 5)]

distances = []

for i in range(len(points)):
    for j in range(i + 1, len(points)): 
        point1 = points[i]
        point2 = points[j]
        
        dis = euclideanDistance(point1, point2)
        mesafe = euclideanDistance(points[i], points[j])
        distances.append(mesafe)


min_distance = min(distances)

print("Tüm noktalar arasındaki mesafeler:", distances)
print("Minimum mesafe:", min_distance)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Verilen noktalar

points = [(1, 2), (4, 6), (7, 8), (2, 1), (9, 5)]


# Mesafeyi hesaplayan fonksiyon
def calculate_distance(p1, p2):
    return np.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)

# Noktaları ve mesafeleri çizmek
plt.figure(figsize=(8, 8))
for i, (x, y) in enumerate(points):
    plt.scatter(x, y, label=f'Point {i+1} ({x}, {y})')
    plt.text(x, y + 0.2, f'({x}, {y})', fontsize=9)

# Mesafeleri çizmek
for i in range(len(points)):
    for j in range(i + 1, len(points)):
        x1, y1 = points[i]
        x2, y2 = points[j]
        mid_x, mid_y = (x1 + x2) / 2, (y1 + y2) / 2
        distance = calculate_distance(points[i], points[j])
        plt.text(mid_x, mid_y, f'{distance:.2f}', fontsize=8, color='red')
        plt.plot([x1, x2], [y1, y2], 'k-', alpha=0.5)

# Grafik ayarları
plt.title("Points and Distances")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```
